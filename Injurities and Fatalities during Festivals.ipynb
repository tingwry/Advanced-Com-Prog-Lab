### Question 1

````python
# Code and result the can answer the question. 
df2 = df[df['รถผู้บาดเจ็บ'] == 'จักรยานยนต์'].pivot_table(index = 'จังหวัด', columns='รถผู้บาดเจ็บ', values='อายุ', aggfunc='count')
print(df2['จักรยานยนต์'].max())
print(df2['จักรยานยนต์'].idxmax())
````

### Question 1 
Compare the number of accidents caused by each type of transportation in ชลบุรี and กรุงเทพมหานคร

````python
df_chonburi = df[df['จังหวัด'] == 'ชลบุรี'].pivot_table(index = 'รถผู้บาดเจ็บ', columns='จังหวัด', values='อายุ', aggfunc='count')
df_chonburi.plot()

df_bkk = df[df['จังหวัด'] == 'กรุงเทพมหานคร'].pivot_table(index = 'รถผู้บาดเจ็บ', columns='จังหวัด', values='อายุ', aggfunc='count')
df_bkk.plot()
````

### Question 2 
Compare the number of different cases of death in เลย and แพร่

````python
df_q2 = df[((df['จังหวัด']=='เลย') | (df['จังหวัด']=='แพร่')) & (df['ผลการรักษา'] != 'ทุเลา/หาย')].pivot_table(index='ผลการรักษา', columns='จังหวัด', values='อายุ', aggfunc='count').fillna(0)

df_q2.plot()
````

### Question 3
Which gender has more death cases?

Answer: Male

````python
df_gender = df[(df['ผลการรักษา'] != 'ทุเลา/หาย')].pivot_table(index='เพศ', columns='ผลการรักษา', values='อายุ', aggfunc='count').fillna(0)
df_gender['deaths'] = df_gender.sum(axis=1)
df_gender = df_gender[['deaths']]
df_gender.plot(kind='bar')
````
